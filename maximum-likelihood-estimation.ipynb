```python
import numpy as np
from scipy.optimize import minimize
```

```python
# LL function for the gaussian distribution
def log_likelihood(params, data):
    mean, var = params
    n = len(data)

    log_likelihood = -0.5 * n * np.log(2 * np.pi * var) \
                     - np.sum((data - mean) ** 2) / (2 * var)

    return - log_likelihood

data = np.array([85, 90, 88, 92, 95])

initial_params = [np.mean(data), np.var(data)]

result = minimize(log_likelihood, initial_params, args=(data,), method="BFGS")

estimated_mean, estimated_var = result.x

print("Initial Params:", initial_params)
print("Estimated Mean:", estimated_mean)
print("Estimated Variance:", estimated_var)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the Gaussian distribution
mean = 90
std_dev = 4

# Generate data points for the x-axis
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)

# Compute the corresponding probabilities using the Gaussian distribution
pdf = norm.pdf(x, mean, std_dev)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Gaussian Distribution')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')

# Add a vertical line at the mean
plt.axvline(x=mean, color='r', linestyle='--', label='Mean')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
